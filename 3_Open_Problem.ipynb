```python
import pickle
import igraph

import pandas as pd

nodes = pd.read_csv("data/ue_25m_1750_N.csv")
edges = pd.read_csv("data/ue_25m_1750_E.csv")

net = open("file_net", 'rb') 
net = pickle.load(net)
net
```

```python
#prendo i link che contengono un ht e un tweet

htindex = [edges.iloc[i,1][0]=='#' for i in range(len(edges))]
edgesht = edges[htindex]
#edgesht.head()
```

```python
#creazione grafo proiezione

add=[]

for i in range(len(edgesht)):
        tweet=edgesht.iloc[i,0]
        for k in range(len(edgesht)):
            if edgesht.iloc[k,0] == tweet and edgesht.iloc[k,1]!=edgesht.iloc[i,1]:
                print(edgesht.iloc[i,1],edgesht.iloc[i,0],edgesht.iloc[k,0],edgesht.iloc[k,1], )
                add.append([edgesht.iloc[i,1], edgesht.iloc[k,1]])

add
```

```python
import pickle 


file_netht = open("file_netht", 'wb') 
pickle.dump(add, file_netht)
file_netht.close()
```
